```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# Download necessary NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')  # Adding this for the missing resource

# Define a function to clean the tweet text
def clean_tweet(tweet):
    # Remove URLs, mentions, hashtags
    tweet = ' '.join([word for word in tweet.split() if 'http' not in word and '@' not in word and '#' not in word])
    
    # Tokenize the tweet and remove stopwords
    words = word_tokenize(tweet)
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word.lower() not in stop_words]
    
    # Return the cleaned tweet as a string
    return ' '.join(filtered_words)

# Apply the cleaning function to the tweet column
df['cleaned_tweet'] = df['Text'].apply(clean_tweet)

# Display the cleaned tweets
df[['Text', 'cleaned_tweet']].head()
```

```python
# Plot the sentiment distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Sentiment')
plt.title('Sentiment Distribution of Tweets')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()
```
